Download the magic04.data data file from the UCI ML Repository. The dataset has 10 real attributes, and the last one is simply the class label, which is categorical, and which you will ignore for this assignment. Assume that attributes are numbered starting from 0.

Write a script to answer the following questions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/magic04.data', header=None)
df.columns = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
df.head()
```

```python
df = df.drop('class', axis=1)
df.head()
```

```python
df.info()
```

## Task 1

```python
multi_mean = df.mean()
multi_mean
```

## Task 2

```python
centered_df = df - multi_mean
centered_df.head()
```

```python
n, d = df.shape
n
```

```python
inner_prod = (1/n) * (centered_df.T @ centered_df)
inner_prod
```

## Task 3

```python
outer_prod = centered_df.iloc[0:1].T @ centered_df.iloc[0:1]
for i in range(1, n):
  outer_prod += centered_df.iloc[i:i+1].T @ centered_df.iloc[i:i+1]

outer_prod /= n
outer_prod
```

## Task 4

```python
from numpy.linalg import norm
```

```python
# cos = (A.B) / (||A||.||B||)

A = centered_df.iloc[:, 1]
B = centered_df.iloc[:, 2]

cosine = (A @ B) / (norm(A) * norm(B))

cosine
```

```python
plt.scatter(A, B)
plt.xlabel(df.columns[1])
plt.ylabel(df.columns[2])
plt.title('Scatter Plot between 2 attributes')
plt.show()
```

## Task 5

```python
mean = df.iloc[:, 1].mean()
mean
```

```python
std = df.iloc[:, 1].std()
std
```

```python
var = df.iloc[:, 1].var()
var
```

```python
from scipy.stats import norm
```

```python
Xax = np.arange(-50, 100, 1)
Yax = norm.pdf(Xax, mean, std)
plt.plot(Xax, Yax)
plt.axvline(mean, color='r', linestyle='dashed')
plt.xlabel('X')
plt.ylabel('PDF')
plt.show()
```
